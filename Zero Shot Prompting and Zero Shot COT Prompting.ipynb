```python
#Zero Shot Prompting 
from langchain_aws import ChatBedrockConverse 
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser 
from langchain_core.runnables import RunnablePassthrough
```

```python
llm = ChatBedrockConverse(model = "cohere.command-r-plus-v1:0", temperature = 0.2, max_tokens = 400)
```

```python
question = """Henry made two stops before he reached his 60 miles bike ride.
                He first stopped after 20 miles and his second stop was 15 miles before the end of the trip. 
                How many miles did he travel between his first and second stop?"""
```

```python
llm.invoke(question).content
```

```python
###Zero shot CoT Prompting 
template = """Question: {question} 
            Based on the given question, provide a step by step solution mention the numbers of steps taken in solving the given problem and provide the calculation steps"""
```

```python
prompt = PromptTemplate(template = template, input_variables = ['question']) 
llm_chain = prompt | llm | StrOutputParser()
```

```python
from pprint import pprint
res = llm_chain.invoke(question)
pprint(res)
```
