```python
from langchain_aws import ChatBedrockConverse
def model_invoke(temp,prompt):
	llm=ChatBedrockConverse(model="anthropic.claude-3-sonnet-20240229-v1:0", #'cohere.command-r-plus-v1:0', #amazon.nova-lite-v1:0
	aws_access_key_id='',
	aws_secret_access_key='',
	temperature=temp,
	region_name='us-east-1',
	max_tokens=50)
	output=llm.invoke(prompt)
	print(f"\n=== Temperature = {temp} ===")
	print(output.content)
	return output.content
```

```python
#Prompt
prompt = "The best way to learn coding is"
#Different temperatures to test
temperatures = [0, 0.2, 0.5, 1]
#Generate responses for each temperature
for temp in temperatures:
	model_invoke(temp,prompt)
```
