```python
from langchain_aws import ChatBedrockConverse 
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory  
#cohere.command-r-plus-v1:0
```

```python
llm=ChatBedrockConverse( 
    model='cohere.command-r-plus-v1:0', 
    region_name='us-east-1', 
    max_tokens=200 
) 

llm.invoke('hi').content
```

```python
chat_history={} 
chat_history
```

```python
def get_session_history(session_id: str) ->InMemoryChatMessageHistory: 
    history=chat_history.get(session_id) 
    if history is None: 
        history=InMemoryChatMessageHistory() 
        chat_history[session_id]=history 
    return history       
```

```python
prompt=ChatPromptTemplate.from_messages([ 
    ("system","You are a helpful assistant."), 
    MessagesPlaceholder(variable_name="chat_history"), 
    ("human","{input}"), 
])
```

```python
runnable=prompt|llm
```

```python
chain=RunnableWithMessageHistory( 
    runnable, 
    get_session_history, 
    input_messages_key="input", 
    history_messages_key="chat_history" 
)
    
```

```python
response1=chain.invoke({"input":"Hi there!"}, config={"configurable":{"session_id":"test_session"}}) 
print("Response 1:",response1) 
print("_______"*20)                       
                       
```

```python
response2=chain.invoke({"input":"I live in Mysore and I love to eat dosa"}, config={"configurable":{"session_id":"test_session"}}) 
print("Response 2:",response2) 
print("_______"*20)     
```
